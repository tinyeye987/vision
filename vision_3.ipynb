```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = 'girl.jpg'
original_image = cv2.imread(image_path)
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB) #for plotting
plt.imshow(original_image_rgb)
plt.title('Original Image (RGB)')
plt.axis('off')
plt.show()
(h, w) = original_image.shape[:2]
print(f"Height: {h}")
print(f"Width: {w}")
center = (w // 2, h // 2)
print(f"Center: {center}")
```

```python
M_90 = cv2.getRotationMatrix2D(center, 90, 1.0)
rotated_90 = cv2.warpAffine(original_image, M_90, (w, h))
rotated_90_rgb = cv2.cvtColor(rotated_90, cv2.COLOR_BGR2RGB)
plt.imshow(rotated_90_rgb)
plt.title('Rotated 90 Degrees')
plt.axis('off')
plt.show()
```

```python
M_45 = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_45 = cv2.warpAffine(original_image, M_45, (w, h))
rotated_45_rgb = cv2.cvtColor(rotated_45, cv2.COLOR_BGR2RGB)
plt.imshow(rotated_45_rgb)
plt.title('Rotated 45 Degrees')
plt.axis('off')
plt.show()
```

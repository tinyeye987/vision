```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'cat.jpg'
original_image = cv2.imread(image_path)
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
plt.imshow(original_image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
def pixelate_by_average(image, block_size):
    # get dimensions of the original image
    (h, w) = image.shape[:2]
    
    # make a copy of the image to store the result
    output_image = image.copy()

    # go through the image: step size = block size
    for y in range(0, h, block_size):
        for x in range(0, w, block_size):
            # handle edges
            y_end = min(y + block_size, h)
            x_end = min(x + block_size, w)
            
            # get the block from original image
            block = image[y:y_end, x:x_end]
            
            #calculate avg and convert it in to int
            avg_color = np.mean(block, axis=(0, 1)).astype(int)
            
            # fill the corresponding block in output image
            output_image[y:y_end, x:x_end] = avg_color
            
    return output_image
```

```python
pixelated_3x3 = pixelate_by_average(original_image_rgb, 3)
plt.imshow(pixelated_3x3)
plt.title('3x3 Block Average')
plt.axis('off')
plt.show()
```

```python
pixelated_5x5 = pixelate_by_average(original_image_rgb, 5)
plt.imshow(pixelated_5x5)
plt.title('5x5 Block Average')
plt.axis('off')
plt.show()
```

```python
pixelated_7x7 = pixelate_by_average(original_image_rgb, 20)
plt.imshow(pixelated_7x7)
plt.title('7x7 Block Average')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'dog.jpg'
original_image = cv2.imread(image_path)
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
plt.imshow(original_image_rgb)
plt.title('Original Image (RGB)')
plt.axis('off')
plt.show()
```

```python
kernel_size_3x3 = (3, 3)
blurred_3x3 = cv2.blur(original_image_rgb, kernel_size_3x3)
plt.imshow(blurred_3x3)
plt.title('3x3 Average')
plt.axis('off')
plt.show()
```

```python
kernel_size_10x10 = (10, 10)
blurred_10x10 = cv2.blur(original_image_rgb, kernel_size_10x10)
plt.imshow(blurred_10x10)
plt.title('10x10 Average')
plt.axis('off')
plt.show()
```

```python
kernel_size_20x20 = (20, 20)
blurred_20x20 = cv2.blur(original_image_rgb, kernel_size_20x20)
plt.imshow(blurred_20x20)
plt.title('20x20 Average')
plt.axis('off')
plt.show()
```

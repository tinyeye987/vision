```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image_path = 'girl.jpg'
img_original = Image.open(image_path)
plt.imshow(img_original)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
if img is None:
    raise FileNotFoundError("Image not found. Check the filename and path.")
```

```python
# Show the original image
plt.imshow(img, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')
plt.show()
```

```python
unique_vals = np.unique(img)
print(f"Number of unique intensity levels: {len(unique_vals)}")
```

```python
def reduce_intensity_levels(img, num_levels):
    # input must be power of 2 between 2 and 256
    if not (2 <= num_levels <= 256 and (num_levels & (num_levels - 1)) == 0):
        raise ValueError("Number of levels must be a power of 2 between 2 and 256.")
    
    # Calculate level size
    level_size = 256 // num_levels

    # Quantize image
    reduced_img = (img // level_size) * level_size
    return reduced_img
```

```python
num_levels = 4
```

```python
# Apply reduction
reduced_img = reduce_intensity_levels(img, num_levels)

# Show reduced image
plt.imshow(reduced_img, cmap='gray')
plt.title(f'Image with {num_levels} Intensity Levels')
plt.axis('off')
plt.show()
unique_vals = np.unique(reduced_img)
print(f"Number of unique intensity levels: {len(unique_vals)}")
```
